```python
#input file
fin = open("rato1.csv", "rt")
#output file to write the result to
fout = open("out1.csv", "wt")
#for each line in the input file
for line in fin:
    fout.write(line.replace('\\n', ',').replace('\'','').replace('\\b', '').replace('b\\r', '').replace('b', '').replace('\\', ',').replace('\n','').replace(' ', ''))
#close input and output files
fin.close()
fout.close()
```

```python
fin = open("out1.csv", "rt")
fout = open("step1.csv", "wt")
#fix date
for line in fin:
    counter = 0
    while counter < len(line):
        #year
        for idx in range(counter, len(line)):
            if line[idx] == '-':
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #month
        for idx in range(counter, len(line)):
            if line[idx] == '-':
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #day
        for idx in range(counter, counter + 2):
            fout.write(line[idx])
            counter = counter + 1
        fout.write(' ')
        #hour
        for idx in range(counter, len(line)):
            if(line[idx] == ':'):
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #minute
        for idx in range(counter, len(line)):
            if(line[idx] == ':'):
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #second
        for idx in range(counter, len(line)):
            if(line[idx] == '.'):
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #fraction of second
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            if line[idx].isnumeric():
                fout.write(line[idx])
            else:
                fout.write(',')
                break
            counter = counter + 1
        #sensor_id
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #sensor_id value
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            if line[idx].isnumeric():
                fout.write(line[idx])
            else:
                fout.write(',')
                break
            counter = counter + 1
        #port
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #port value
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            if line[idx].isnumeric():
                fout.write(line[idx])
            else:
                fout.write(',')
                break
            counter = counter + 1
        #axes
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #axes 1
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                fout.write(',')
                counter = counter + 1
                break
            if line[idx].isnumeric() or line[idx] == '-':
                fout.write(line[idx])
            else:
                fout.write(',')
                break
            counter = counter + 1
        
        #axes 2
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            if line[idx].isnumeric() or line[idx] == '-':
                fout.write(line[idx])
            else:
                fout.write(',')
                break
            counter = counter + 1
        #axes 3
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            if line[idx].isnumeric() or line[idx] == '-':
                fout.write(line[idx])
            else:
                fout.write(',')
                break
            counter = counter + 1
        #temperature
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #temperature value
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write('\n')
                break
            fout.write(line[idx])
            counter = counter + 1
fin.close()
fout.close()
```

```python
file = open("step2.csv", "rt")
fout = open("step3.csv", "wt")
counter = 1
fout.write("time,sensor_id,port,axes_x,axes_y,axes_z,temperature\n")
for line in file:
    if(counter == 1):
        counter = counter + 1
        continue
    linearray = line.split(',')
    linearray[0] = datetime.datetime.strptime(linearray[0], '%Y-%m-%d %H:%M:%S.%f').timestamp()
    asdasdasd = str(linearray[0]).replace('.', '')
    asdasdasd = int(asdasdasd)
    fout.write(','.join([str(asdasdasd), linearray[1], linearray[2], linearray[3], linearray[4], linearray[5], linearray[6]]))
    counter = counter + 1
file.close()
fout.close()
```
