```python
import csv
import datetime
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import numpy as np
import seaborn as sns
import pandas
from mpl_toolkits.mplot3d import Axes3D
from sklearn import preprocessing
```

```python
#input file
fin = open("rato1.csv", "rt")
#output file to write the result to
fout = open("out1.csv", "wt")
#for each line in the input file
for line in fin:
    fout.write(line.replace('\\n', ',').replace('\'','').replace('\\b', '').replace('b\\r', '').replace('b', '').replace('\\', ',').replace('\n','').replace(' ', ''))
#close input and output files
fin.close()
fout.close()
```

```python
fin = open("out1.csv", "rt")
fout = open("step1.csv", "wt")
#fix date
for line in fin:
    counter = 0
    while counter < len(line):
        #year
        for idx in range(counter, len(line)):
            if line[idx] == '-':
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #month
        for idx in range(counter, len(line)):
            if line[idx] == '-':
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #day
        for idx in range(counter, counter + 2):
            fout.write(line[idx])
            counter = counter + 1
        fout.write(' ')
        #hour
        for idx in range(counter, len(line)):
            if(line[idx] == ':'):
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #minute
        for idx in range(counter, len(line)):
            if(line[idx] == ':'):
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #second
        for idx in range(counter, len(line)):
            if(line[idx] == '.'):
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #fraction of second
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            if line[idx].isnumeric():
                fout.write(line[idx])
            else:
                fout.write(',')
                break
            counter = counter + 1
        #sensor_id
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #sensor_id value
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            if line[idx].isnumeric():
                fout.write(line[idx])
            else:
                fout.write(',')
                break
            counter = counter + 1
        #port
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #port value
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            if line[idx].isnumeric():
                fout.write(line[idx])
            else:
                fout.write(',')
                break
            counter = counter + 1
        #axes
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #axes 1
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                fout.write(',')
                counter = counter + 1
                break
            if line[idx].isnumeric() or line[idx] == '-':
                fout.write(line[idx])
            else:
                fout.write(',')
                break
            counter = counter + 1
        
        #axes 2
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            if line[idx].isnumeric() or line[idx] == '-':
                fout.write(line[idx])
            else:
                fout.write(',')
                break
            counter = counter + 1
        #axes 3
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            if line[idx].isnumeric() or line[idx] == '-':
                fout.write(line[idx])
            else:
                fout.write(',')
                break
            counter = counter + 1
        #temperature
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write(line[idx])
                break
            fout.write(line[idx])
            counter = counter + 1
        #temperature value
        for idx in range(counter, len(line)):
            if(line[idx] == ','):
                counter = counter + 1
                fout.write('\n')
                break
            fout.write(line[idx])
            counter = counter + 1
fin.close()
fout.close()
```

```python
def clean_vector(vectors, target_length):
    cleaned_vectors = []
    seconds = vectors[0][0]
    sum_vector = [0,0,0]
    mean_vector = [0,0,0]
    counter = 0
    for vector in vectors:
        if vector[0] < 0:
            print(vector)
            continue
        if vector[0] != seconds:
            mean_vector = [sum_vector[0] / counter, sum_vector[1] / counter, sum_vector[2] / counter]
            cleaned_vectors.append([seconds, mean_vector[0],mean_vector[1],mean_vector[2]])
            counter = 0
            seconds = vector[0]
            sum_vector = [0,0,0]
        sum_vector = [sum_vector[0] + vector[1],sum_vector[1] + vector[2],sum_vector[2] + vector[3]]
        counter = counter + 1
    return cleaned_vectors
```

```python
def magnitude_of_vectors(vectors):
    dot_product = []
    for vector in vectors:
        dot_product.append([vector[0],np.linalg.norm(vector[1:])])
    return dot_product
```

```python
def unify(vectors1, vectors2):
    unified = []
    x = range(len(vectors1))
    for i in x:
        unified.append([(vectors1[i][0]+vectors2[i][0])/2, (vectors1[i][1]+vectors2[i][1])/2])
    return unified
```

```python
def plot_graph_v(vector, titulo, outlier, treshold):
    vectordf = pandas.DataFrame(vector, columns=['time', 'x','y', 'z'])
    if(outlier):
        vectordf = vectordf[(np.abs(stats.zscore(vectordf.x)) < treshold)]
        vectordf = vectordf[(np.abs(stats.zscore(vectordf.y)) < treshold)]
        vectordf = vectordf[(np.abs(stats.zscore(vectordf.z)) < treshold)]
    plt.figure(num=None, figsize=(18, 6), dpi=80, facecolor='w', edgecolor='k')
    fig, ax1 = plt.subplots()
    fig.set_figheight(6)
    fig.set_figwidth(18)
    ax1.plot(vectordf.time, vectordf.x, label='x')
    ax1.plot(vectordf.time, vectordf.y, label='y')
    ax1.plot(vectordf.time, vectordf.z, label='z')
    ax2 = ax1.twinx()
    ax2.set_ylabel('Frequencia (s)', color='tab:red')
    ax2.plot(df_buxco.time[0:], df_buxco.frequency[0:], label='frequencia', color='tab:red')
    ax1.set_title(titulo)
    ax1.set_xlabel('Tempo (s)')
    ax1.legend(loc=2)
    ax2.legend(loc=1)
    fig.savefig(titulo + '.png')
```

```python
plot_graph_v(cl_vectors_29, "ACELEROMETRO ID 29 E FREQUENCIA", True, 1.2)
plot_graph_v(cl_vectors_30, "ACELEROMETRO ID 30 E FREQUENCIA", True, 3)
plot_graph_v(cl_vectors_105, "ACELEROMETRO ID 105 E FREQUENCIA", True, 4)
```

```python
def plot_graph_m(vector, titulo, outlier, treshold):
    vectordf = pandas.DataFrame(vector, columns=['time', 'magnitude'])
    if(outlier):
        vectordf = vectordf[(np.abs((vectordf.magnitude - vectordf.magnitude.mean())/vectordf.magnitude.std(ddof=0)) < treshold)]
    fig, ax1 = plt.subplots()
    plt.figure(num=None, figsize=(18, 6), dpi=80, facecolor='w', edgecolor='k')
    fig.set_figheight(6)
    fig.set_figwidth(18)
    ax1.plot(vectordf.time, vectordf.magnitude, label='x')
    ax1.set_title(titulo)
    ax1.set_xlabel('Tempo (s)')
    ax2 = ax1.twinx()
    ax2.set_ylabel('Frequencia', color='tab:orange')
    ax2.plot(df_buxco.time[0:], df_buxco.frequency[0:], label='frequency', color='tab:orange')
    fig.savefig(titulo + '.png')
```
